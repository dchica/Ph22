```python
from matplotlib import pyplot as plt
import numpy as np
import sys
```

```python
#This function allows me to interpret the input as mathematical function
def parse(f):
    '''This method takes the function f(t) as a string input'''
    ''' and converts it into a mathematical function.'''
    if isinstance(f, str) == False:
        sys.exit('Please input f(t) as a string as you would using numpy')
    flist = f.split('t');
    fchange = ''
    ffinal = []
    for i, e in enumerate(flist):
        if i == len(flist) - 1:
            fchange += e
        else:
            fchange += e
            fchange += '%s'
    ffinal.append(fchange)
    ffinal.append(len(flist) - 1)
    return ffinal

def fEuler(oldf, derf, t, h):
    '''This method takes a function f(t), its derivative f'(t), a value for t, and a stepsize h.'''
    '''It uses the forward Euler method to find an approximate value for f(t+h) and it returns the'''
    '''approximation, and the error of this approximation when compared to the actual value of f(t+h).'''
    func, n_len = parse(oldf);
    dfunc, dn_len = parse(derf);
    tn = (str(t),) * n_len;
    td = (str(t),) * dn_len;
    th = (str(t + h),) * n_len;
    
    f1 = eval(func % tn)
    f2 = eval(dfunc % td)
    
    newf = f1 + h * f2
    error = np.abs(eval(func % th) - newf)
    return(newf, error)

def midpoint(oldf, t, h):
    '''This method takes a function f(t), a value for t, and a stepsize h.'''
    '''It uses the Midpoint method to find an approximate value for f(t+h) and it returns the'''
    '''approximation, and the error of this approximation when compared to the actual value of f(t+h).'''
    func, n_len = parse(oldf);
    tn = (str(t),) * n_len;
    th = (str(t + h),) * n_len;
    
    f = eval(func % tn)
    ft = (str(f),) * n_len;
    fz = eval(func % ft)
    
    zbar = f + (h / 2) * fz
    zbar_t = (str(zbar),) * n_len;
    fzbar = eval(func % zbar_t)
    
    newf = f + h * fzbar
    
    error = np.abs(eval(func % th) - newf)
    return(newf, error)
```

```python
#Here I am iterating the methods 10 times for h=.3
n = 0;
fe_list = [];
mp_list = [];
fe_i = 5;
mp_i = 5;
while n < 10:
    val_fe, err_fe = fEuler('.1 * t**2', '.1 * 2*t', fe_i, .3);
    val_mp, err_mp = midpoint('.1 * t**2', mp_i, .3);
    fe_list.append(err_fe);
    mp_list.append(err_mp);
    fe_i = val_fe;
    mp_i = val_mp;
    n += 1;
```

```python
#Here I am plotting the error as we iterate the method
plt.plot(fe_list, label = 'Forward Euler')
plt.plot(mp_list, label = 'Midpoint Method')
plt.legend()

plt.xlabel("Number of iterations");
plt.ylabel("Error");

plt.savefig('Error1.png')
```

Problem 3

```python
#Here I am returning the value of the errors when we change the value of h
feh_list = [];
mph_list = [];
h_list = [];
h = 2;
while h > (2 / 16):
    err_fe = fEuler('.1 * t**2', '.1 * 2*t', 5, h)[1];
    err_mp = midpoint('.1 * t**2', 5, h)[1];
    feh_list.append(err_fe);
    mph_list.append(err_mp);
    h *= .5;
    h_list.append(h * 2);
```

```python
#Here I am plotting the error as we change the value of h
plt.plot(h_list, feh_list, label = 'Forward Euler')
plt.plot(h_list, mph_list, label = 'Midpoint Method')
plt.legend()

plt.xlabel("Number of iterations");
plt.ylabel("Error");

axes = plt.gca();
axes.invert_xaxis()

plt.savefig('Error1.png')
```

```python
def derf(f, t, h):
    '''This method takes a function f at t and returns'''
    '''an approximation of its derivative at that t'''
    func, n_len = parse(f);
    tp = (str(t + h),) * n_len;
    tm = (str(t - h),) * n_len;
    
    fp = func % tp;
    fm = func % tm;
    
    return ('(%s - %s) / (2* %s)' % (fp, fm, h))

def RK4(oldf, derf, t, h):
    '''This method takes a function f(t), a function that finds the derivative f'(t), a value for t, and a stepsize h.'''
    '''It uses the Runge-Kutta method to find an approximate value for f(t+h) and it returns the'''
    '''approximation, and the error of this approximation when compared to the actual value of f(t+h).'''
    func, n_len = parse(oldf);
    tn = (str(t),) * n_len;
    th = (str(t + h),) * n_len;
    
    k_l = [];
    k_l.append(('(%s *' % h) + derf(oldf, t, h) +')');
    k_l.append(('(%s *' % h) + derf(oldf +'+ .5 *'+ k_l[0], t + .5 * h, h) +')');
    k_l.append(('(%s *' % h) + derf(oldf +'+ .5 *'+ k_l[1], t + .5 * h, h) +')');
    k_l.append(('(%s *' % h) + derf(oldf +'+'+ k_l[2], t + h, h) +')');
    
    vl = [];
    for i in k_l:
        vl.append(eval(i))
        
    newf = eval(func % tn) + (1/6)*(vl[0] + vl[1] + vl[2] + vl[3]);
    error = np.abs(eval(func % th) - newf)
    return(newf, error)
```

Problem 5

Since we want a circular orbit, the radius must remain constant (unchanging with respect to time) so let's set it to R=1. Since we expect circular orbit, I define x=Cos(t), y=Sin(t) and run the method on both.

```python
x_list = []
y_list = []
t_list = np.linspace(0, 1000, num=1001);

for i in t_list:
    x_list.append(RK4('np.cos(t)', derf, i, .001)[0])
    y_list.append(RK4('np.sin(t)', derf, i, .001)[0])
```

```python
plt.plot(x_list, y_list)
plt.gca().set_aspect('equal', adjustable='box')
plt.savefig('RKorbit.png')
```
